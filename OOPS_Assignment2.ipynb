Q18. What is the "diamond problem" in multiple inheritance, and how does Python address it?


Ans-18)The "diamond problem" occurs in multiple inheritance when a subclass inherits from two classes that have a common ancestor. If both parent classes have a method with the same name, ambiguity arises about which method to use.

Python resolves the diamond problem using a method resolution order (MRO). It follows a specific order (C3 Linearization) to determine the order in which methods are resolved, avoiding ambiguity.
